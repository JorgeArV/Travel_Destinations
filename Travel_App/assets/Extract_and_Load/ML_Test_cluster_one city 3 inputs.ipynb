```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import os
import pickle
```

```python
df = pd.read_csv(os.path.join("NorthAmericaCityAirbnb.csv"))
```

```python
df
```

```python
df['Price'] = df['Price'].str.replace('$','')
df['Price'] = df['Price'].str.replace(',','')
df['Price'] = pd.to_numeric(df['Price'])
```

```python
df.dropna(subset = ["Reviews Scores Rating"], inplace=True)
```

```python
df.dtypes
```

# Neighbors

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
df
```

```python
df_filtered = df[(df.City == "Montreal,QC")]
df_filtered
```

```python
df_dummies = pd.get_dummies(df_filtered, columns=["Room Type"])
df_dummies.head(100)
```

```python
#from keras.utils import to_categorical
```

```python
#y_train_categorical = to_categorical(encoded_y_train)
```

```python
# Neighbors - 3 inputs: Price, Rating and room type
```

```python
X= df_dummies[["Price", "Reviews Scores Rating", "Room Type_Entire home/apt"]]
y = df_dummies["Neighbourhood"]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train_scaled, y_train)
model.score(X_test_scaled, y_test)
```

```python
train_score = []
test_score = []
for k in range(1, 20, 2): 
  model = KNeighborsClassifier(n_neighbors=k)
  model.fit(X_train_scaled, y_train)
  train_score.append(model.score(X_train_scaled, y_train))
  test_score.append(model.score(X_test_scaled, y_test))
  print(k, model.score(X_train_scaled, y_train), model.score(X_test_scaled, y_test))
```

```python
data_point = [[100, 90, 0]]
model.predict(data_point)
```
