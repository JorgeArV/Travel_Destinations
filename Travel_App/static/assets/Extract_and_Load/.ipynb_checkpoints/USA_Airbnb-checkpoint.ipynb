```python
import pandas as pd
```

```python
Asheville = pd.read_csv('../Resources/Asheville_NorthCarolina.csv')
```

```python
Asheville.rename(columns = {
    'host_name':'Host Name',
    'neighbourhood': 'Neighbourhood',
    'latitude': 'Latitude',
    'longitude': 'Longitude',
    'room_type': 'Room Type',
    'price': 'Price',
    'minimum_nights': 'Minimum Nights',
    'availability_365': 'Availability 365'}, inplace = True)
```

```python
Asheville["City"] = "Ashville,NC"
```

```python
Asheville = Asheville[['City','Host Name','Neighbourhood','Latitude','Longitude','Room Type','Price','Minimum Nights','Availability 365']]
Asheville.head()
```
