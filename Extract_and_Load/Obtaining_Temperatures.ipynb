```python
# 1.Importing dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
import os
```

```python
#2. We add url. Each city will have a new url

# Jorge ####  url2 = "https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/Washington%20DC/2020-1-1/2020-12-31?unitGroup=metric&key=XXXXXXXXXXXXXXXXXX&include=obs"

# marina ### 
url2 = "https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/Porto/2020-1-1/2020-12-31?unitGroup=metric&key=XXXXXXXXXXXXX&include=obs"
```

```python
response = requests.get(url2).json()
```

```python
response
```

```python
#3. We create empty lists to transfer the data
temperature = []
temp_max = []
temp_min = []
humidity = []
precip = []
snow = []
windspeed = []
conditions = []
date = []
city = []
```

```python
#4. We loop thropugh the API and extract the info we need
for x in range(366):
    temperature.append(response["days"][x]["temp"])
    temp_max.append(response["days"][x]["tempmax"])
    temp_min.append(response["days"][x]["tempmin"])
    humidity.append(response["days"][x]["humidity"])
    precip.append(response["days"][x]["precip"])
    snow.append(response["days"][x]["snow"])
    windspeed.append(response["days"][x]["windspeed"])
    conditions.append(response["days"][x]["conditions"])
    date.append(response["days"][x]["datetime"])
    city.append("Porto, PT")
```

```python
print(len(temperature),len(temp_max),len(temp_min),len(humidity),len(precip),len(snow),len(windspeed),len(conditions),len(date),len(city))
```

```python
#5. We create a dataframe based on the lists
New_df = pd.DataFrame(
    {'City': city,
     'Date': date,
     'Temperature': temperature,
     'temp_max': temp_max,
     'temp_min':temp_min,
     'humidity':humidity,
     'precip':precip,
     'snow':snow,
     'windspeed':windspeed,
     'conditions': conditions
    })
```
