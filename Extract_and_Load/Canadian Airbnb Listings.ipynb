```python
MontrealOriginal = pd.read_csv('Montreal.csv')
MontrealOriginal.head()
```

```python
MontrealOriginal.rename(columns = {
    'host_name':'Host Name',
    'neighbourhood': 'Neighbourhood',
    'latitude': 'Latitude',
    'longitude': 'Longitude',
    'room_type': 'Room Type',
    'price': 'Price',
    'minimum_nights': 'Minimum Nights',
    'availability_365': 'Availability 365'}, inplace = True)
MontrealOriginal.head()
```
