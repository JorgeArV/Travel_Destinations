```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join("Temperature_Dataset1.csv"))
```

```python
df
```

```python
newdf = df[df.City.isin(["Quebec City, QC", "New Brunswick, NB", "Twin Cities, MN", "Ottawa, ON", "Montreal, QC", "Hawaii, HI", "Broward County, FL", "New Orleans, LA", "Austin, TX", "Clark County, NV"])]

newdf
```

```python
cluster = []

for i in df['Date']:
    if i < "2020-09-01" and i > "2020-04-01": 
        cluster.append("1")
    else:
        cluster.append("2")
        
df["cluster"] = cluster
print(df)
```

```python
cluster1 = []

for i in df['Temperature']:
    if i < -10: 
        cluster1.append("1")
    elif i > -9 and i < 0:
        cluster1.append("2")
    elif i >1 and i < 15:
        cluster1.append("3")
    elif i > 16 and i <25:
        cluster1.append("4")
    else:
        cluster1.append("5")
        
df["cluster1"] = cluster1
print(df)
```

```python
y = newdf["City"]
```

```python
X= newdf[["temp_max", "temp_min", "humidity"]]
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
predictions = model.predict(X_test)
```
