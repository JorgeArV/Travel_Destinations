```python
import pandas as pd
```

```python
NewYorkOriginal = pd.read_csv('New York.csv')
NewYorkOriginal.head()
```

```python
NewYorkOriginal.rename(columns = {
    'host_name':'Host Name',
    'neighbourhood': 'Neighbourhood',
    'latitude': 'Latitude',
    'longitude': 'Longitude',
    'room_type': 'Room Type',
    'price': 'Price',
    'minimum_nights': 'Minimum Nights',
    'availability_365': 'Availability 365'}, inplace = True)
NewYorkOriginal.head()
```

```python
NewYorkOriginal["City"] = "New York City,NY"
```

```python
NewYork = NewYorkOriginal[['City','Host Name','Neighbourhood','Latitude','Longitude','Room Type','Price','Minimum Nights','Availability 365']]
NewYork.head()
```
